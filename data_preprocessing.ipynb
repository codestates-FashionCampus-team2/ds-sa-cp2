## 5. 금액 관련 컬럼
- 컬럼 설명
    - total_amount : 총 비용
    - product_id : 상품id
    - quantity : 주문 수량
    - item_price : 상품 가격

- 전처리 결과 <br>
: "상품 가격 x 상품 수량 - 할인 가격 + 배송료 = 총 비용"에 맞는 데이터와 맞지 않는 데이터 분리 예정

```python
# "상품 가격 x 상품 수량 - 할인 가격 + 배송료 = 총 비용"이 맞지 않는 데이터 확인
df_ver2[df_ver2['quantity']*df_ver2['item_price']+df_ver2['shipment_fee']-df_ver2['promo_amount'] != df_ver2['total_amount']]
```

```python
# "상품 가격 x 상품 수량 - 할인 가격 + 배송료 = 총 비용"이 맞는 데이터 확인
df_ver2[df_ver2['quantity']*df_ver2['item_price']+df_ver2['shipment_fee']-df_ver2['promo_amount'] == df_ver2['total_amount']]
```

## 6. 상품 정보 관련 컬럼
- 컬럼 설명
    - id : 상품 id
    - gender_x : 상품 타겟 성별
    - masterCategory : 최상위 분류
    - subCategory : 서브 분류
    - articleType : 상품 종류
    - baseColour : 기본 색상 (결측치 O)
    - season : 적합 계절 (결측치 O)
    - year : 상품 출시 년도 (결측치 O)
    - usage : 사용 방식 (결측치 O)
    - productDisplayName : 상품 이름 (결측치 O)

- 전처리 결과 <br>
    - 해당 컬럼이 결측값이더라도 다른 컬럼을 사용할 수 있고, 결측치가 많지 않아(제일 많은 컬럼이 8871개) 결측치 데이터를 제거하지 않고 EDA 진행 → 후에 모델링 진행시에 결측치 처리 예정
    - id 컬럼은 product_id 컬럼과 같기 때문에 drop
    - 상품 타겟 성별 컬럼은 "gender_x → product_gender"로 변경
    - 기본 색상 컬럼은 "baseColour → baseColor"로 변경

```python
df_ver2.drop(columns=['id'], inplace=True)
df_ver3 = df_ver2.rename(columns={'baseColour':'baseColor', 'gender_x':'product_gender'})
```
